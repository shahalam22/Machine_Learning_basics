<h2>Opening csv file from a URL</h2>

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36 Edg/127.0.0.0"}
req = requests.get(url, headers = headers)
data = StringIO(req.text)

pd.read_csv(data)
```

```python
def rename(name):
    if name == "Rasmussen Group":
        return "RG"
    else:
        return name
```

```python
# pd.read_csv('customers-100.csv').head(1)
pd.read_csv('customers-100.csv',converters={'Company':rename}).head(1)
```

```python
pd.read_csv('customers-100.csv',na_values=['abir'])

# ekhane na_values=[keywords] te keywords jegulo thakbe era shob NaN value hoye jabe
```

```python
dfs = pd.read_csv('customers-100.csv', chunksize=15)

# chunksize pura data set ke chunk e divide kore. So, kono kaj korte gele for loop e shob chunk chalaye kaj kore fela jabe

for chunks in dfs:
    print(chunks.shape)
```
