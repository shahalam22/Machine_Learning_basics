```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
```

```python
x = 6 * np.random.rand(200,1) - 3
y = 0.8 * x**2 + 0.9 * x + 2 + np.random.randn(200,1)

# y = 0.8x^2 + 0.9x + 2
```

```python
plt.plot(x,y,'b.')
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

### Applying Linear Regression

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
plt.plot(x_train,lr.predict(x_train),color='r')
plt.plot(x,y,'b.')
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### Appplying Polynomial Regression

```python
poly = PolynomialFeatures(degree=2)  # include_bias=False korle bias parameter er value ashbe na. Tokhon power joto toto gulo result ashbe. x^0 er value ta bad jabe. x^0 ta holo bias factor

x_train_trans = poly.fit_transform(x_train)
x_test_trans = poly.transform(x_test)
```

```python
print(x_train[0])
print(x_train_trans[0])  # jehetu power 2 tai 3 ta x0,x1,x2 ashbe transform korle
```

```python
lr = LinearRegression()
lr.fit(x_train_trans,y_train)
```

```python
y_pred = lr.predict(x_test_trans)
```

```python
r2_score(y_test,y_pred)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
x_new = np.linspace(-3,3,200).reshape(200,1)
x_new_poly = poly.transform(x_new)
y_new = lr.predict(x_new_poly)
```

```python
plt.plot(x_new,y_new,"r-",linewidth=2,label="Predictions")
plt.plot(x_train,y_train,"b.",label="Training points")
plt.plot(x_test,y_test,"g.",label="Testing points")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
