```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
df['species'] = encoder.fit_transform(df['species'])
```

```python
df.head()
```

```python
new_df = df[df['species']!=0][['sepal_width','petal_length','species']]
```

```python
new_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.scatter(new_df['sepal_width'],new_df['petal_length'],c=new_df['species'],cmap='winter')
```

```python
# Taking only 10 rows for training
new_df2 = new_df.sample(100)
new_train_df2 = new_df.iloc[:60,:].sample(10)
new_val_df2 = new_df.iloc[60:80,:].sample(5)
new_test_df2 = new_df.iloc[80:,:].sample(5)
```

```python
new_train_df2
```

```python
x_test = new_val_df2.iloc[:,0:2].values
y_test = new_val_df2.iloc[:,-1].values
```

## Case 1 - Bagging

```python
# Data for Tree 1
df_bag = new_train_df2.sample(8,replace=True)

x = df_bag.iloc[:,0:2]
y = df_bag.iloc[:,-1]

df_bag
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from mlxtend.plotting import plot_decision_regions
from sklearn.metrics import accuracy_score
```

```python
def evaluate(clf,x,y):
    clf.fit(x,y)
    plot_tree(clf)
    plt.show()
    plot_decision_regions(x.values,y.values,clf=clf, legend=2)
    y_pred = clf.predict(x_test)
    print(accuracy_score(y_test,y_pred))
```

```python
dt_bag1 = DecisionTreeClassifier()
evaluate(dt_bag1,x,y)
```

```python
# Data for Tree 2
df_bag = new_train_df2.sample(8,replace=True)

x = df_bag.iloc[:,0:2]
y = df_bag.iloc[:,-1]

df_bag
```

```python
dt_bag2 = DecisionTreeClassifier()
evaluate(dt_bag2,x,y)
```

```python
# Data for Tree 3
df_bag = new_train_df2.sample(8,replace=True)

x = df_bag.iloc[:,0:2]
y = df_bag.iloc[:,-1]

df_bag
```

```python
dt_bag3 = DecisionTreeClassifier()
evaluate(dt_bag3,x,y)
```
