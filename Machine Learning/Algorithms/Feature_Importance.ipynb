## How Feature Importance is Calculated in Decision Tree

We will see the calculation of feature importance in Decision tree then will see how it happens in Random Forest. As we know Random Forest is a collection of Decision Tree

```python
from sklearn.datasets import make_classification
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x,y = make_classification(n_samples=15,n_classes=2,n_features=2,n_informative=2,n_redundant=0,random_state=0)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(x,y)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
```

```python
clf.feature_importances_
```

```python
# now in Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=2)
rf.fit(x,y)
```

```python
rf.feature_importances_
```

```python
print(rf.estimators_[0].feature_importances_)
print(rf.estimators_[1].feature_importances_)
```
