### Creating Function to perform RIDGE Regression through GRADIENT DESCENT

```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import numpy as np
```

```python
x,y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
reg = SGDRegressor(penalty='l2',max_iter=500,eta0=0.1,learning_rate='constant',alpha=0.001)
```

```python
reg.fit(x_train,y_train)

y_pred = reg.predict(x_test)
print("R2_score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=0.001, max_iter=500,solver='sparse_cg')
```

```python
reg.fit(x_train,y_train)

y_pred = reg.predict(x_test)
print("R2_score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

### Creating our own RIDGE REGRESSION class (Gradient Descent)

```python
class MeraRidgeGD:
    
    def __init__(self,learning_rate,epochs,alpha=0.1):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.alpha = alpha
        self.m = None
        self.b = None

    def fit(self,x_train,y_train):

        self.coef_ = np.ones(x_train.shape[1])
        self.intercept_ = 0
        thetha = np.insert(self.coef_,0,self.intercept_)

        x_train = np.insert(x_train,0,1,axis=1)

        for i in range(self.epochs):
            thetha_der = np.dot(x_train.T,x_train).dot(thetha) - np.dot(x_train.T, y_train) - self.alpha*thetha
            thetha = thetha - self.learning_rate*thetha_der

        self.coef_ = thetha[1:]
        self.intercept_ = thetha[0]
        
    def predict(self,x_test):
        return np.dot(x_test,self.coef_) + self.intercept_
```

```python
reg = MeraRidgeGD(epochs=500,alpha=0.001,learning_rate=0.005)
```

```python
reg.fit(x_train,y_train)

y_pred = reg.predict(x_test)
print("R2_score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

```python
class MeraRidge:
    def __init__(self,alpha=0.1):
        self.alpha = alpha
        self.m = None
        self.b = None

    def fit(self,x_train,y_train):
        x_train = np.insert(x_train,0,1,axis=1)
        I = np.identity(x_train.shape[1])
        I[0][0] = 0  # ekhane Beta0(intercept) er jonno I matrix er value ) dhore kaj hoise
        result = np.linalg.inv(np.dot(x_train.T,x_train) + self.alpha * I).dot(x_train.T).dot(y_train)
        self.intercept_ = result[0]
        self.coef_ = result[1:]

    def predict(self, x_test):
        return np.dot(x_test,self.coef_) + self.intercept_
```

```python
reg = MeraRidge()
reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
print(r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```
