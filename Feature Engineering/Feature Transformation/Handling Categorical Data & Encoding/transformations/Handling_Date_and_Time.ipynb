```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# Extract year
date['date_year'] = date['date'].dt.year
```

```python
date.head()
```

```python
# Extract month
date['date_month'] = date['date'].dt.month
date.head()
```

```python
# Extract month's name
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

```python
# Extract date
date['date_day'] = date['date'].dt.day
date.head()
```

```python
# Day of the week
date['date_dow'] = date['date'].dt.dayofweek
date.head()
```

```python
# Day of the week - name
date['date_dow_name'] = date['date'].dt.day_name()
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# Is weekend
date['date_is_wekend'] = np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# Extract week of the year
date['date_week'] = date['date'].dt.isocalendar().week
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# Extract Quarter
date['quater'] = date['date'].dt.quarter
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# Extract Semester
date['semester'] = np.where(date['quater'].isin([1,2]),1,2)
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# Extract Time elapsed between dates
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# week passed 
np.round((today-date['date']) / np.timedelta64(1,'W'),0)
```
